```python
import os, sys
print(os.getcwd())
sys.path.append(os.chdir(os.getcwd()+"/.."))


# simple Example plot Enveloped Potential with two Harmonic Oscilators
##Imports:
import matplotlib.pyplot as plt
import math
import numpy as np
import ConveyorBelt.src.potential2D as pot
import ConveyorBelt.visualisation.plotPotentials as exPlot
%matplotlib inline
```

```python
#definitions
#Potentials
positions = np.linspace(-180, 180, 360)#1 degree steps
V1 = pot.wavePotential2D(phase_shift=(0,0), multiplicity=(1.0, 1.0), amplitude=(1.0, 1.0))
V2 = pot.wavePotential2D(phase_shift=(0.0, 0.0), multiplicity=(1.0, 1.0), amplitude=(1.0, 1.0))

#calc
positions2D = [(x, y) for x in positions for y in positions]
energies_V1 = V1.ene(positions2D)
energies_V2 = V2.ene(positions2D)
```

```python
resolution  =len(positions)
print(len(energies_V1))
energies_V1_Map = [[row for row in energies_V1[x:x+resolution]] for x in range(0,len(energies_V1), resolution)]
print(len(energies_V1_Map), len(energies_V1_Map[0]))
energies_V2_Map = [[row for row in energies_V2[x:x+resolution]] for x in range(0,len(energies_V2), resolution)]
energy_map = [np.add(row1, row2) for row1, row2 in zip(energies_V1_Map, energies_V2_Map)]

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=[15,10])
surf = ax1.imshow(energies_V1_Map, cmap="inferno", interpolation="nearest", origin='center', vmax=4, vmin=-4, extent=[min(positions), max(positions), min(positions), max(positions)])
surf = ax2.imshow(energies_V2_Map, cmap="inferno", interpolation="nearest", origin='center', vmax=4, vmin=-4,extent=[min(positions), max(positions), min(positions), max(positions)])
surf = ax3.imshow(energy_map, cmap="inferno", interpolation="nearest", origin='center', vmax=4, vmin=-4,extent=[min(positions), max(positions), min(positions), max(positions)])
fig.colorbar(surf,fraction=0.046, pad=0.04)
#fig.savefig("HUGEMAGE.png")

```
